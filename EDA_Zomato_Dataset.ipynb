```python
# read another dataset Country code
df1=pd.read_excel("/content/Country-Code.xlsx")
df1.head()
```

```python
# can we combine two data set Zomato and country code(df and df1)
# ans yes
# using pandas merge function
final_df=pd.merge(df,df1,on="Country Code",how="left")
final_df.head()
```

```python
# to check data type
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
print(country_names)
```

```python
country_value=final_df.Country.value_counts().values
print(country_value)
```

```python
# PIE chart
plt.pie(country_value,labels=country_names)

# its output mess up all
```

```python
# pie chart to top 3 country that uses zomato
plt.pie(country_value[:3],labels=country_names[:3],autopct="%1.2f%%")   # aotopct - auto percentage
```

```python
# rename 0 column to reating count
Ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating counts"})
Ratings

# Ratings is important feature
```

# Observation

1.When rating is betwwen 4.5 to 4.9----->Excellent

2.When rating is between 4.0 to 4.4------>Very Good

3.when rating is between 3.5 to 3.9----->Good

4.when rating is between 2.5 to 3.4----->Average

5.when rating is between 1.8 to 2.4------>Poor

```python
Ratings.head()
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"]=(12,6)      # it use becoz barplot figure size will increase
sns.barplot(data=Ratings,x="Aggregate rating",y="Rating counts")
```

```python
sns.barplot(data=Ratings,x="Aggregate rating",y="Rating counts",hue="Rating color")
# checkrating color not match with barplot white color showing blue ,red show orange so on
# we want to map the color
```

```python
# we want to map the color
#here
sns.barplot(data=Ratings,x="Aggregate rating",y="Rating counts",hue="Rating color",palette=['blue','red','orange','yellow','green','darkgreen'])
#white change to blue becoz white not see
```

# Obersevation
1.Not Rated is vary high.

2.Maximum rating is between 2.5 to 3.4.

```python
# count plot----> for categorical columns used it
sns.countplot(x="Rating color",data=Ratings,palette=['blue','red','orange','yellow','green','darkgreen'])
```
